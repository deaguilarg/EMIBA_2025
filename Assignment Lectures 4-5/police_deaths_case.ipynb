```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('police_deaths_USA_v6.csv')
```

```python
df
```

```python
plt.figure(figsize=(12, 12))  # Width = 12, Height = 6
ax = sns.barplot(df.groupby('State').count()[['Rank']].sort_values('Rank', ascending=False), x="Rank", y="State",legend=False)
ax.bar_label(ax.containers[0], fontsize=10);
plt.show()
```

```python
plt.figure(figsize=(12, 12))  # Width = 12, Height = 6
ax = sns.barplot(df.groupby('Cause').count()[['Rank']].sort_values('Rank', ascending=False), x="Rank", y="Cause",legend=False)
ax.bar_label(ax.containers[0], fontsize=10);
plt.show()
```

```python
plt.figure(figsize=(12, 12))  # Width = 12, Height = 6
ax = sns.barplot(df.groupby('Day_Of_Week').count()[['Rank']].sort_values('Rank', ascending=False), x="Day_Of_Week", y="Rank",legend=False)
ax.bar_label(ax.containers[0], fontsize=10);
plt.show()
```
